```kotlin
import mdk.gsm.builder.buildWalker
import mdk.gsm.util.StringVertex
import mdk.gsm.util.GsmDotGenerator
import mdk.gsm.state.ITransitionGuardState
import mdk.gsm.graph.Graph

import guru.nidi.graphviz.engine.Format
import guru.nidi.graphviz.engine.Graphviz
import mdk.gsm.builder.buildGraphOnly
import mdk.gsm.state.NoTransitionGuardState
import java.io.File

val one = StringVertex("1")
val two = StringVertex("2")
val three = StringVertex("3")
val four = StringVertex("4")
val five = StringVertex("5")
val six = StringVertex("6")
val seven = StringVertex("7")
val eight = StringVertex("8")

val graph = buildGraphOnly<StringVertex, String, NoTransitionGuardState, Nothing> {

        addVertex(one) {
            addEdge {
                setTo(two)
            }
            addEdge {
                setTo(three)
            }
        }

        addVertex(two) {
            addEdge {
                setTo(four)
            }
        }

        addVertex(three) {
            addEdge {
                setTo(five)
            }
            addEdge {
                setTo(six)
            }
        }

        addVertex(four) {
            addEdge {
                setTo(eight)
            }
        }

        addVertex(five) {
            addEdge {
                setTo(seven)
            }
        }

        addVertex(six) {
            addEdge {
                setTo(seven)
            }
        }

        addVertex(seven) {
            addEdge {
                setTo(eight)
            }
        }

        addVertex(eight)

}


val dotGenerator = GsmDotGenerator<StringVertex, String, NoTransitionGuardState, Nothing>()
val dotContent = dotGenerator.generateDot(graph, "EightVertexDAG")
val dotFile = File("diagram.dot")
dotFile.writeText(dotContent)

val svg = Graphviz.fromString(dotContent)
    .render(Format.SVG).toString()

HTML(svg)
```
